```python
import pandas as pd
import ast
from typing import List
```

```python
lts_llm_res = pd.read_csv('../../llm_output/llm_res_df.csv')
mapped_vars = pd.read_csv('../../llm_output/mapped_vars_df.csv')
```

```python
lts_llm_res['llm_response'] = lts_llm_res['llm_response'].apply(lambda x: ast.literal_eval(x)[1])
```

```python
def get_llm_alginment(response):
    for i in response:
        if i.startswith("ALIGNMENT: "):
            
            return i
```

```python
def join_lines(lines: List[str]) -> str:
    return "\n".join(lines)
```

```python
user_curation = pd.DataFrame(
    columns=[
        "user_variable",
        "user_description",
        "matched_text",
        "uid",
        "relevance_score",
        "llm_alignment",
        "match",
    ]
)
```

```python

user_variable = []
user_description = []
matched_text = []
uid = []
relevance_score = []
llm_alignment = []
match = []
```

```python
r_count = 1

for index, row in mapped_vars.iterrows():

    matching_row = lts_llm_res[lts_llm_res["user_variable"] == row["user_variable"]]
    alignment = ""
    if r_count == 1:
        print(matching_row["llm_response"])
        for _, matching_row in matching_row.iterrows():
            alignment = matching_row["llm_response"]
            break
    if r_count == 3:
        r_count = 1
    else:
        r_count += 1

    user_variable.append(row["user_variable"])
    user_description.append(row["user_description"])
    matched_text.append(row["matched_text"])
    uid.append(row["uid"])
    relevance_score.append(row["relevance_score"])
    llm_alignment.append(alignment)
    match.append("")

user_curation["user_variable"] = user_variable
user_curation["user_description"] = user_description
user_curation["matched_text"] = matched_text
user_curation["uid"] = uid
user_curation["relevance_score"] = relevance_score
user_curation["llm_alignment"] = llm_alignment
user_curation["match"] = match

    
```
